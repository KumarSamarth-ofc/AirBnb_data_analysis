#Handling missng values

```python
df['last review'] = pd.to_datetime(df['last review'] , errors = 'coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
df.fillna({'reviews per month':0,'last review':df['last review'].min()},inplace=True)
```

```python
df.dropna(subset = ['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns =['license','house_rules'],errors = 'ignore')
```

```python
df.head(

)
```

#Data type of price and service is object with $ so changing to integer

```python
df['price'] = df['price'].replace('[\$,]','', regex=True).astype(float)
```

```python
df['service fee'] = df['service fee'].replace('[\$,]','', regex=True).astype(float)
```

```python
df.head()
```

#removing duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

#descriptive stats

```python
df.describe()
```

#Visualization

##What is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins = 50 , kde = True, color = 'red')
plt.title('Distribution of Listing Prices')
plt.xlabel('Price in $')
plt.ylabel('Frequency')
plt.show();
```

COnclusion - The histogram is fairly evenly distributed of listing prices across different price ranges and KDE is showing clearly that there is no much fluctuations in price of variety of products

#
How are listings distributed across different neighborhoods?

```python
plt.figure(figsize=(12,8))
sns.countplot(y = 'neighbourhood group', data = df, color = 'lightblue', order=df['neighbourhood group'].value_counts().index )
plt.title("	No. of Listings by Neighbourhood Group")
plt.xlabel('Count ')
plt.ylabel('	Neighbourhood group')
plt.show()
```

#
What is the relationship between price and room type?

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'room type', y= 'price' ,hue='room type' , data=df, palette='Set1')
plt.title("Price vs Room type")
plt.xlabel('Room type')
plt.ylabel('Price in $')
plt.show()
```

#
How has the number of reviews changed over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time  = df.groupby(df['last review'].dt.to_period('M')).size()
reviews_over_time = reviews_over_time[reviews_over_time.index.year <= 2024]
```

```python
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
